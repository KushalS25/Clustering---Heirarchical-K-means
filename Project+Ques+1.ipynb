```python
# Univariate Analysis

def univariateAnalysis_numeric(column):
    print("Description of " + column)
    print("----------------------------------------------------------------------------")
    print(df[column].describe(),end=' ')
    
    plt.figure()
    print("Distribution of " + column)
    print("----------------------------------------------------------------------------")
    sns.displot(df[column], kde=True,color='pink');
    plt.show()
    
    plt.figure()
    print("BoxPlot of " + column)
    print("----------------------------------------------------------------------------")
    ax = sns.boxplot(x=df[column],color='lightblue')
    plt.show()
```

```python
for x in df.columns:
    univariateAnalysis_numeric(x)
```

```python
fig, ax = plt.subplots(3, 3, figsize = (16, 10))
ax = ax.flatten()

for i, c in enumerate(df.columns):
    sns.boxplot(x = df[c], ax = ax[i],color='pink')
plt.suptitle('Outlier Analysis using BoxPlots', fontsize = 12)
fig.tight_layout()
```

**As we see 'probability_of_full_payment' & 'min_payment_amt' have outliers**

```python
# Finding Outlier for probability_of_full_payment Enteries 
IQR =stats.iqr(df.probability_of_full_payment, nan_policy='omit')
IQR
```

```python
## Calculation of Q1 and Q3
Q1 = df.probability_of_full_payment.quantile(0.25)
Q3 = df.probability_of_full_payment.quantile(0.75)
print(Q1,Q3)
```

```python
## Calculation of Upper and Lower limits
lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
print(lower_limit,upper_limit)
```

```python
# List of Outliers in probability_of_full_payment column
df[(df.probability_of_full_payment<lower_limit)|(df.probability_of_full_payment>upper_limit)]
```

```python
df.probability_of_full_payment.min()
```

```python
df.probability_of_full_payment.max()
```

```python
df.probability_of_full_payment.mean()
```

```python
df.probability_of_full_payment.median()
```

```python
# Treating The Outliers with flooring
df['probability_of_full_payment'] = np.where(df['probability_of_full_payment']<lower_limit,lower_limit,df['probability_of_full_payment'])
```

```python
# Checking whether the change is affected or not
df[(df.probability_of_full_payment<lower_limit)|(df.probability_of_full_payment>upper_limit)]
```

```python
# Checking OUTLiers for min_Payment_amt
## Calculation of Q1 and Q3
Q1 = df.min_payment_amt.quantile(0.25)
Q3 = df.min_payment_amt.quantile(0.75)
print(Q1,Q3)
```

```python
IQR =stats.iqr(df.min_payment_amt, nan_policy='omit')
IQR
```

```python
## Calculation of Upper and Lower limits
lower_limit = Q1-1.5*IQR
upper_limit = Q3+1.5*IQR
print(lower_limit,upper_limit)
```

```python
# List of Outliers in probability_of_full_payment column
df[(df.min_payment_amt<lower_limit)|(df.min_payment_amt>upper_limit)]
```

```python
# Treating Outliers with Capping
df['min_payment_amt'] = np.where(df['min_payment_amt']>upper_limit,upper_limit,df['min_payment_amt'])
```

```python
df[(df.min_payment_amt<lower_limit)|(df.min_payment_amt>upper_limit)]
```

```python
# Checking Boxplots to see if the Outliers are treated successfully

fig, ax = plt.subplots(3, 3, figsize = (16, 10))
ax = ax.flatten()

for i, c in enumerate(df.columns):
    sns.boxplot(x = df[c], ax = ax[i],color='lightblue')
plt.suptitle('Outlier Analysis using BoxPlots', fontsize = 12)
fig.tight_layout()
```

## Normalizing and Scaling


```python
#Scales the data. Essentially returns the z-scores of every attribute
from scipy.stats import zscore
df_scaled=df.apply(zscore)
df_scaled.head()
```

```python
df_scaled.describe().T
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
#Using Average Linkage

avg_link = linkage(df_scaled, method='average')
```

```python
avg_link
```

```python
dend = dendrogram(avg_link)
```

```python
dend = dendrogram(avg_link,truncate_mode='lastp',p=30)
```

```python
dend = dendrogram(avg_link,truncate_mode='lastp',p=20)
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
clusters_m=fcluster(avg_link,3,criterion='maxclust')
```

```python
clusters_m
```

```python
clusters_d=fcluster(avg_link,2.9,criterion='distance')
```

```python
clusters_d
```

```python
np.array_equal(clusters_m,clusters_d)
```

```python
df['Cluster']=clusters_m
```

```python
df.head()
```

```python
df.to_csv('clustered_bank.csv')
```

```python
k_means = KMeans(n_clusters = 3)
k_means.fit(df_scaled)
labels = k_means.labels_
```

```python
df["Clus_kmeans"] = labels
df.head(5)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
silhouette_score(df_scaled,labels)
```

```python
sil_width = silhouette_samples(df_scaled,labels)
```

```python
df["sil_width"] = sil_width
df.head(5)
```

```python
silhouette_samples(df_scaled,labels).min()
```
